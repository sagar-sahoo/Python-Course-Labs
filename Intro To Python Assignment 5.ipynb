```python
import pandas as pd

planes = pd.read_csv('planes.csv')
flights = pd.read_csv('flights.csv')
```

```python
flights['dep_delay'].mean()
```

```python
flights.groupby(['carrier'])['dep_delay'].mean().reset_index()
```

```python
flights.groupby(['carrier']).agg({'dep_delay':['mean', 'min', 'max','median']})
```

```python
flights.groupby(['carrier']).agg({'dep_delay':['mean'],'air_time':['median']})
```

```python
flights.groupby(['origin','dest']).agg({'carrier':['count']}).reset_index().sort_values(by=2,axis=1)
```

```python
routes_df = flights.groupby(['origin','dest'])['carrier'].count().reset_index()
```

```python
routes_df.sort_values(by='carrier',ascending=False)
```
